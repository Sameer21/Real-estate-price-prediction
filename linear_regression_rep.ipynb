```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns
from sklearn import linear_model;
```

```python
df=pd.read_csv('randomized_real_estate_data_csv.md')
df
```

```python
%matplotlib inline
plt.figure(figsize=(10,10))
plt.xlabel('AREA(sqft)')
plt.ylabel('PRICE($)')
sns.scatterplot(df,x='area',y='price',markers='*',palette='viridis',hue='price')
```

```python
rg=linear_model.LinearRegression()
rg.fit(df[['area']],df.price)
```

```python
rg.predict([[2100]])
```

```python
plt.figure(figsize=(10,10))
plt.xlabel('AREA')
plt.ylabel('PRICE')
plt.plot(df.area,rg.predict(df[['area']]),color='blue')
sns.scatterplot(df,x='area',y='price',markers='*',palette='viridis',hue='price')
plt.show()
```
