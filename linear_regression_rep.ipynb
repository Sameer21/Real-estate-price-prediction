```python
"""
Real Estate Price Prediction Model
---------------------------------
This script demonstrates linear regression for predicting real estate prices based on property area.

Author: Sameer
Repository : Public
"""
```

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns
from sklearn import linear_model
```

```python
"""
    Load and prepare the real estate dataset.

    Parameters:
    -----------
    file_path : str
        Path to the CSV file containing real estate data

    Returns:
    --------
    pandas.DataFrame
        Loaded and preprocessed dataframe
    """
df=pd.read_csv('randomized_real_estate_data_csv.md')
```

```python
df
```

```python
"""
    Create scatter plot to visualize relationship between area and price.

    Parameters:
    -----------
    df : pandas.DataFrame
        DataFrame containing 'area' and 'price' columns
"""

%matplotlib inline
sns.set_style("whitegrid")
plt.figure(figsize=(10,10))
plt.xlabel('AREA(sqft)')
plt.ylabel('PRICE($)')
sns.scatterplot(df,x='area',y='price',markers='*',palette='viridis',hue='price')
```

```python
 """
    Train a linear regression model to predict real estate prices.

    Parameters:
    -----------
    df : pandas.DataFrame
        DataFrame containing 'area' and 'price' columns

    Returns:
    --------
    sklearn.linear_model.LinearRegression
        Trained linear regression model
"""
    
# Initialize the linear regression model
rg=linear_model.LinearRegression()
# Train the model
rg.fit(df[['area']],df.price)
```

```python
"""
    Predict the price for a given property area.

    Parameters:
    -----------
    model : sklearn.linear_model.LinearRegression
        Trained linear regression model
    area : float
        Property area in sqft

    Returns:
    --------
    float
        Predicted property price
    """

rg.predict([[2100]])
```

```python
"""
    Visualize the data points and the regression line.

    Parameters:
    -----------
    df : pandas.DataFrame
        DataFrame containing 'area' and 'price' columns
    model : sklearn.linear_model.LinearRegression
        Trained linear regression model
"""



plt.figure(figsize=(10,10))
plt.xlabel('AREA')
plt.ylabel('PRICE')
plt.plot(df.area,rg.predict(df[['area']]),color='blue')
sns.scatterplot(df,x='area',y='price',markers='*',palette='viridis',hue='price')
plt.show()
```
